```python
def get_200_800_data(data):
    ans = []
    start_low = 0
    start_high = 0
    #0
    list_proccess_name = list(df[df['process_name']==0].index)
    list_proccess_name.append(29783)
#     print(list_proccess_name)
    list_high_height = list(df[(df['height']>860)&(df['height']<920)].index)  #41
    list_low_height = list(df[(df['height']>=170)&(df['height']<201)].index)  #53
    print('low {}'.format(list_low_height))
    print('high {}'.format(list_high_height))

    for i in range(len(list_proccess_name)-1):
        left = list_proccess_name[i];right = list_proccess_name[i+1]-1
        #800
        high_temp = -1
        for j in range(start_high,len(list_high_height)):
            if list_high_height[j] <=right:
                high_temp = list_high_height[j]
            else:
                start_high = j
                break
        low_temp = -1
        for k in range(start_low,len(list_low_height)):
            if list_low_height[k]<=right:
                if low_temp==-1:
                    low_temp = list_low_height[k]
            else:
                start_low = k
                break
        if low_temp ==-1 and high_temp !=-1:
            print('{}--{}'.format(left,right))
        if low_temp !=-1 and high_temp ==-1:
            print('{}--{}'.format(left,right))
        if low_temp ==-1 and high_temp ==-1:
            print('{}--{}'.format(left,right))
        ans.append([low_temp,high_temp])
    return ans 
    
```

```python
index_new = get_200_800_data(df)
print(len(index_new))
print(index_new)
```
