```python
def get_start_index(data):
    ans = [0]
    temp = -1
    for index, row in data.iterrows():
        if int(row["process_name"])<temp:
            ans.append(index)
        temp = int(row["process_name"])
        
    return ans

ans = get_start_index(df)
print(ans)
```

```python
def get_train_val_test_list(v_rate,t_rate):
    ans = get_start_index(df)
    ans.append(len(df)-1)
    train_list = [] ;train_nums = 0
    val_list = [];val_nums = 0
    test_list = [];test_nums = 0
    for i in range(len(ans)-1):
        st = ans[i];ed = ans[i+1]
        differ = ed - st
        
        train_nums += int(differ*v_rate)
        train_list.append([st,st+int(differ*v_rate)])
        
        val_nums += int(differ*t_rate) - int(differ*v_rate)
        val_list.append([st+int(differ*v_rate),st+int(differ*t_rate)])
        
        test_nums +=  differ - int(differ*t_rate)
        test_list.append([st+int(differ*t_rate),ed])
#         print(train_nums,val_nums,test_nums)
    return train_list,val_list,test_list,train_nums,val_nums,test_nums
```

```python
train_list,val_list,test_list,train_nums,val_nums,test_nums = get_train_val_test_list(0.75,0.95)
```

```python
def generator(data, lookback, delay, list_index,
              shuffle=False, batch_size=128, step=6):
    for index_1 in list_index:
        min_index = index_1[0]
        max_index =index_1[1]
        if max_index is None:
            max_index = len(data) - delay - 1
        i = min_index + lookback
        while 1:
            if shuffle:
                rows = np.random.randint(
                    min_index + lookback, max_index, size=batch_size)
            else:
                if i + batch_size >= max_index:
                    i = min_index + lookback
                rows = np.arange(i, min(i + batch_size, max_index))
                i += len(rows)

            samples = np.zeros((len(rows),
                               lookback // step,
                              data.shape[-1]))
            targets = np.zeros((len(rows)))
            for j, row in enumerate(rows):
                indices = range(rows[j] - lookback, rows[j], step)
                samples[j] = data[indices]
                targets[j] = data[rows[j] + delay][0]
            yield samples, targets
```

```python
lookback = 20
step = 1
delay = 2
batch_size = 64

train_gen = generator(float_data,
                      lookback=lookback,
                      delay=delay,
                      list_index = train_list,
                      shuffle=True,
                      step=step, 
                      batch_size=batch_size)
val_gen = generator(float_data,
                    lookback=lookback,
                    delay=delay,
                  list_index = val_list,
                  shuffle=True,
                    step=step,
                    batch_size=batch_size)
test_gen = generator(float_data,
                     lookback=lookback,
                     delay=delay,
                     shuffle=True,
                    list_index = test_list,
                     step=step,
                     batch_size=batch_size)

train_steps = train_nums//batch_size
# This is how many steps to draw from `val_gen`
# in order to see the whole validation set:
val_steps = (val_nums - lookback) // batch_size

# This is how many steps to draw from `test_gen`
# in order to see the whole test set:
test_steps = (test_nums - lookback) // batch_size
```
