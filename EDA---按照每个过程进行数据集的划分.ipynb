```python
df = pd.read_csv('./export_data/negative_direction/fillNA-mean/5-5/my_data_clean.csv')
df = df.drop('Unnamed: 0',axis=1)
df = df.sort_index(axis=1)
df
```

```python
def get_start_index(data):
    ans = [0]
    temp = -1
    for index, row in data.iterrows():
        if int(row["process_name"])<temp:
            ans.append(index)
        temp = int(row["process_name"])
        
    return ans

ans = get_start_index(df)
print(ans)
```

```python
def get_train_val_test_list(v_rate,t_rate):
    ans = get_start_index(df)
    ans.append(len(df)-1)
    train_list = [] ;train_nums = 0
    val_list = [];val_nums = 0
    test_list = [];test_nums = 0
    for i in range(len(ans)-1):
        st = ans[i];ed = ans[i+1]
        differ = ed - st
        
        train_nums += int(differ*v_rate)
        train_list.append([st,st+int(differ*v_rate)])
        
        val_nums += int(differ*t_rate) - int(differ*v_rate)
        val_list.append([st+int(differ*v_rate),st+int(differ*t_rate)])
        
        test_nums +=  differ - int(differ*t_rate)
        test_list.append([st+int(differ*t_rate),ed])
#         print(train_nums,val_nums,test_nums)
    return train_list,val_list,test_list,train_nums,val_nums,test_nums
```

```python
train_list,val_list,test_list,train_nums,val_nums,test_nums = get_train_val_test_list(0.75,0.95)
```

```python
def get_train_val_test(df,train_list,val_list,test_list):
    train_frames = []
    val_frames = []
    test_frames = []
    for i in train_list:
        train_frames.append(df.iloc[i[0]:i[1]])
    for i in val_list:
        val_frames.append(df.iloc[i[0]:i[1]])
    for i in test_list:
        test_frames.append(df.iloc[i[0]:i[1]])
    return pd.concat(train_frames),pd.concat(val_frames),pd.concat(test_frames)
```

```python
df_train,df_val,df_test = get_train_val_test(df,train_list,val_list,test_list)
print(df_train.shape,df_val.shape,df_test.shape)
```
